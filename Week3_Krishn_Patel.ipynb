```python
import math
```

```python
# we are provided rhe following values and greeks for this problem are as followed and to be fixed throughout 

S0 = 100
K = 100
r = 0.06
sigma = 0.25
T = 1  
N = 3  
```

```python
# 1 time steps 

dt = T/N 
print(dt)
```

```python
# 2. CRR up and down factor 

u = math.exp(sigma*(math.sqrt(dt)))

print(u)

d = 1/u
print(d)
print(round(u*d, 2))
```

```python
# 3 risk free probablity p 

p = (math.exp(r*dt) - d)/ (u-d)
print(round(p, 4))
```

```python
# 4 discount factor 

df = math.exp(-r* dt)
print(round(df,4))
```

```python
# THREE STEPS STOCK PRICE TREE 

# S(N,J)= S0*(U**N)*(D**J) 

# TIME 0 
print(S0)

# TIME 1 

s11 = S0*u

s10 = S0*d

# TIME 2 

s22 = S0*u*u

s21 = S0*u*d
s12 = S0*d*u

s20 = S0*d*d 

# TIME 3 

s33 = S0*u*u*u 

s32 = S0*u*u*d

s31 = S0*u*d*d

s30 = S0*d*d*d

print(S0 , s11 , s10 , s22 , s21 , s20 , s33 , s32 , s31 , s30)
```

```python
# 5 CALCULATING PAYOFF A

#PAYOFF AT S33 
ps33 = max(S0 - s33 , 0)

#PAYOFF AT S32 
ps32 = max(S0 - s32 , 0)

#PAYOFF AT S31 
ps31 = max(S0 - s31 , 0)

#PAYOFF AT S30 
ps30 = max(S0 - s30 , 0)

print(ps30 , ps31 , ps32 , ps33)
```

```python
# valuation at node 2 

v22 = df* ((p*ps33)+((1-p)*ps32))

v21 = df* ((p*ps32)+((1-p)*ps31))

v20 = df* ((p*ps31)+((1-p)*ps30))

print(v22 , v21 , v20)
```

```python
# valuation on node 1 

v11 = df* ((p*v22)+((1-p)*v21))

v10 = df* ((p*v21)+((1-p)*v20))

print(v11 , v10)
```

```python
# valuation at node 0 

v00 = df* ((p*v11)+((1-p)*v10))

print(v00)
```

```python
# call put options parity 

cp = S0 - (K * math.exp(-r*T))

print(cp)
```

```python
# call price payoff 

cp33 = max(s33 - K , 0)

cp32 = max(s32 - K , 0)

cp31 = max(s31 - K , 0)

cp30 = max(s30 - K , 0)

print(cp33 , cp32 , cp31 , cp30)
```

```python
# CALL VALUATION AT NODE 2 

cv22 = df* ((p*cp33)+((1-p)*cp32))

cv21 = df* ((p*cp32)+((1-p)*cp31))

cv20 = df* ((p*cp31)+((1-p)*cp30))

print(cv22 , cv21 , cv20)




```

```python
# call valuation at node 1 

cv11 = df* ((p*cv22)+((1-p)*cv21))

cv10 = df* ((p*cv21)+((1-p)*cv20))

cv00 = df* ((p*cv11)+((1-p)*cv10))


print(cv11 , cv10 , cv00)
```

```python
# BLACK SCHOLES FORMULA 

d1 = (
    math.log(S0 / K)
    + (r + (sigma**2) / 2) * T
) / (sigma * math.sqrt(T))

d2 = d1 - sigma * math.sqrt(T)

print("d1 =", d1)
print("d2 =", d2)
```

```python
N_d1 = 0.643 
N_d2 = 0.546

call = (S0*N_d1) - (K * (math.exp(-r*T))*N_d2)

print(call)
```

```python
put = call - S0 + (K * math.exp(-r*T))

print(put)
```
